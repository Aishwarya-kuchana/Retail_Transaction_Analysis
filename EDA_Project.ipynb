```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\lenovo\Desktop\EDA project dataset\Retail_Dataset.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(5,2))
sns.boxplot(df["TotalAmount"])
plt.title('TotalAmount Boxplot')
plt.show()
```

```python
sns.histplot(df["TotalAmount"], kde=True, bins=50)
plt.title('TotalAmount Distribution')
plt.show()
```

```python
sns.histplot(df["DiscountApplied(%)"], kde=True, bins=50)
plt.title('DiscountApplied(%) Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="ProductCategory", order=df["ProductCategory"].value_counts().index)
plt.title('Productcategory Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="PaymentMethod", order=df["PaymentMethod"].value_counts().index)
plt.title('PaymentMethod Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="State", order=df["State"].value_counts().index)
plt.title('State Frequency')
plt.xticks(rotation=90)
plt.show()
```
