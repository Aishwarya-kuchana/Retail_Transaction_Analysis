```python
import re

def extract_state(location):
    # Match 2 uppercase letters between comma and ZIP
    match = re.search(r',\s*([A-Z]{2})\s*\d{5}', str(location))
    if match:
        return match.group(1)
    return None

df['State'] = df['StoreLocation'].apply(extract_state)
```

```python
import re

def extract_state(location):
    if pd.isna(location):
        return None
    # Matches 2 uppercase letters before a 5-digit ZIP code
    match = re.search(r'([A-Z]{2})\s*\d{5}', str(location))
    if match:
        return match.group(1)
    return None

df['State'] = df['StoreLocation'].apply(extract_state)
```

# Sales Trend by Month

```python
monthly_sales = df.groupby('Month')['TotalAmount'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])

plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
